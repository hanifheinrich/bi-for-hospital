```python
#Normalisasi Kolom Tanggal, Cara Keluar, Diagnosa
import pandas as pd

# Membaca file CSV
df = pd.read_csv("dataset.csv")

# Menghapus kata setelah "Pindah Ruangan" dan "Rujuk" dalam kolom "Cara Keluar"
df['Cara Keluar'] = df['Cara Keluar'].str.replace(r'(Pindah Ruangan\s).*$', r'\1', regex=True)
df['Cara Keluar'] = df['Cara Keluar'].str.replace(r'(Rujuk\s).*$', r'\1', regex=True)

# Membagi nilai dalam kolom Diagnosa menjadi beberapa kolom baru
diagnosa_split = df['Diagnosa'].str.split(' \+ ', expand=True)

# Menyisipkan kolom baru ke DataFrame
df = pd.concat([df, diagnosa_split], axis=1)

# Menyebut kolom baru yang dihasilkan
kolom_baru = ['Diagnosa' + str(i+1) for i in range(diagnosa_split.shape[1])]

# Menyertakan kolom baru ke DataFrame
df.columns = list(df.columns[:-diagnosa_split.shape[1]]) + kolom_baru

# Menyimpan kembali ke file CSV
df.to_csv("preprocessing1.csv", index=False)
```

```python
#Normalisasi Kolom Diagnosa



# Baca data dari file CSV
df = pd.read_csv('preprocessing1.csv')

# Hapus kolom dari DataFrame
df = df.drop(['Diagnosa', 'Diagnosa5', 'Diagnosa6', 'Diagnosa7','Diagnosa8','Diagnosa9','Diagnosa10','Diagnosa11', 'Diagnosa12','Diagnosa4'], axis=1)

import pandas as pd

# Daftar kolom yang akan dinormalisasi
kolom_diagnosa = ['Diagnosa1', 'Diagnosa2', 'Diagnosa3']

# Iterasi melalui setiap kolom untuk normalisasi
for kolom in kolom_diagnosa:
    # Ubah teks menjadi huruf kecil
    df[kolom] = df[kolom].str.lower()
    # Menghapus karakter '#'
    df[kolom] = df[kolom].str.replace('#', '')
    # Hapus angka dan angka perbandingan
    df[kolom] = df[kolom].str.replace(r'\b\d+/\d+\b', '', regex=True)
    # Menghapus tanda kurung beserta isinya
    df[kolom] = df[kolom].str.replace(r'\(.*?\)', '', regex=True).str.strip()
    # Hapus karakter '+' dan semua karakter setelahnya
    df[kolom] = df[kolom].str.replace(r'\+.*', '', regex=True)
    # Hapus karakter '-' dan semua karakter setelahnya
    df[kolom] = df[kolom].str.replace(r'-.*', '', regex=True)
    # Hapus karakter ',' dan semua karakter setelahnya
    df[kolom] = df[kolom].str.replace(r',.*', '', regex=True)
    # Hapus spasi tambahan di awal dan akhir setiap string
    df[kolom] = df[kolom].str.strip()

df.to_csv('preprocessing2.csv', index=False)
```

```python
#Gabung Kolom Diagnosa1,2,3 untuk Normalisasi

import pandas as pd

# Membaca file CSV
df = pd.read_csv('preprocessing2.csv')

# Menggabungkan kolom 'Diagnosa1', 'Diagnosa2', dan 'Diagnosa3' ke bawah
df_melted = df.melt(id_vars=df.columns.difference(['Diagnosa1', 'Diagnosa2', 'Diagnosa3']),
                    value_vars=['Diagnosa1', 'Diagnosa2', 'Diagnosa3'],
                    var_name='Diagnosa Type',
                    value_name='Diagnosa')


# Menyimpan hasil pengolahan ke dalam file CSV baru
df_melted[['Diagnosa']].to_csv('preprocessing3.csv', index=False)
```

```python
#Memecah Diagnosa yang sudah digabung sebelumnya menjadi 3 kolom

import pandas as pd

# Membaca file CSV yang telah diproses sebelumnya
df = pd.read_csv('preprocessing5.csv', skip_blank_lines=False)

# Menghitung jumlah baris dalam setiap bagian
total_rows = len(df)
per_column = total_rows // 3

# Memisahkan DataFrame menjadi tiga bagian dengan menjaga baris kosong
df_diagnosa1 = df.iloc[:per_column]
df_diagnosa2 = df.iloc[per_column:2*per_column].reset_index(drop=True)  # Menetapkan ulang indeks
df_diagnosa3 = df.iloc[2*per_column:].reset_index(drop=True)  # Menetapkan ulang indeks

# Menggabungkan kembali DataFrame dengan kolom 'Diagnosa1', 'Diagnosa2', dan 'Diagnosa3' dan menamai kolomnya
df_merged = pd.concat([df.drop(columns='Diagnosa'), df_diagnosa1,
                       df_diagnosa2, df_diagnosa3], axis=1)

# Menamai kolom-kolom hasil penggabungan
df_merged.columns = ['Diagnosa1', 'Diagnosa2', 'Diagnosa3']

# Menyimpan data terpisah ke dalam file CSV baru
df_merged.to_csv('preprocessing6.csv', index=False)
```

```python
# Menggabunggkan dengan dataset awal dan peneyesuaian karakteristik data

import pandas as pd

# Baca data dari preprocessing6.csv dan preprocessing2.csv
preprocessing6 = pd.read_csv('preprocessing6.csv')
preprocessing2 = pd.read_csv('preprocessing2.csv')

# Ganti nama kolom menjadi huruf kecil (lowercase)
preprocessing6.columns = preprocessing6.columns.str.lower()
preprocessing2.columns = preprocessing2.columns.str.lower()

# Memulai nilai pada kolom Diagnosa1, Diagnosa2, dan Diagnosa3 dengan huruf kapital
preprocessing6['diagnosa1'] = preprocessing6['diagnosa1'].str.capitalize()
preprocessing6['diagnosa2'] = preprocessing6['diagnosa2'].str.capitalize()
preprocessing6['diagnosa3'] = preprocessing6['diagnosa3'].str.capitalize()

# Ganti nama kolom cara bayar menjadi cara_bayar, cara keluar menjadi cara_keluar, dan asal daerah menjadi asal_daerah
preprocessing2.rename(columns={'cara bayar': 'cara_bayar', 'cara keluar': 'cara_keluar', 'asal daerah': 'asal_daerah'}, inplace=True)

# Ganti kolom Diagnosa1, Diagnosa2, dan Diagnosa3 di preprocessing2.csv
preprocessing2['diagnosa1'] = preprocessing6['diagnosa1']
preprocessing2['diagnosa2'] = preprocessing6['diagnosa2']
preprocessing2['diagnosa3'] = preprocessing6['diagnosa3']

# Simpan hasil perubahan kembali ke preprocessing2.csv
preprocessing2.to_csv('dataetl.csv', index=False)

print("Data berhasil digabungkan dan disimpan dalam preprocessing2.csv")
```
